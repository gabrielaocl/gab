```python
import pandas as pd # for reading csv files

from statsmodels.formula.api import ols #regression analysis imports

from nltk.sentiment.vader import SentimentIntensityAnalyzer as stm #for NLP scores
import nltk
nltk.download('vader_lexicon')
```

```python
from google.colab import files
import os

# Loads CSV file from local computer
if not os.path.exists('stockdailyhlnews.csv'):
  uploaded = files.upload()
  if 'stockdailyhlnews.csv' not in uploaded:
    print("Please upload 'stockdailyhlnews.csv' to proceed.")
  else:
    print("File 'stockdailyhlnews.csv' uploaded successfully.")
else:
  print("File 'stockdailyhlnews.csv' already exists.")

df = pd.read_csv('stockdailyhlnews.csv')

print(df.shape)
print(df.head())
```

```python
#1

sia = stm()

df['sscore'] = df['news'].apply(lambda headline: sia.polarity_scores(str(headline))['compound']) #add column called sscore

print(df[['date', 'news', 'sscore']].head())
```

```python
avgsscore = df['sscore'].mean() #average number stored in svgsscore variable

print('Average sentiment compound score:', avgsscore)
```

```python
# 2)

# OLS regression
model = ols('ibm ~ sscore + sp500', data=df).fit()

print(model.summary())
```

```python
# 2) b
adj_rsquared = model.rsquared_adj

print('Adjusted R-squared:', adj_rsquared) #Store adjusted rsquaures
```

```python
# 2) c
f_pvalue = model.f_pvalue #Store pvalue of f-statistics

print('F-statistic p-value:', f_pvalue)
```

```python
# 2) d
sscore_pvalue = model.pvalues['sscore'] #Store pvalue of sscore

print('sscore p-value:', sscore_pvalue)
```

```python
# 2) e
sp500_pvalue = model.pvalues['sp500'] #Store pvalue of sp500

print('sp500 p-value:', sp500_pvalue)
```

```python
# 2) f determine if relationship with IBM score
sscore_rel = sscore_pvalue < 0.05

print('sscore relationship with IBM:', sscore_rel)
```

```python
# 2) g sp500 determine relationship with IBM
sp500_rel = sp500_pvalue < 0.05

print('sp500 relationship with IBM:', sp500_rel)
```
